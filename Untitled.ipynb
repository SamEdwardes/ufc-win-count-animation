```r
library(tidyverse)
library(gganimate)
```

```r
url <- "https://github.com/SamEdwardes/ufc-data/raw/master/raw_total_fight_data.csv"
df <- read_delim(url, delim = ";", col_type = cols()) %>%
    janitor::clean_names() %>%
    mutate(date = lubridate::mdy(date))

glimpse(df)
```

```r
winner_df <- df %>%
    filter(winner != "NA") %>%
    mutate(month = lubridate::month(date)) %>%
    select(winner, date) %>%
    group_by(winner) %>%
    arrange(winner, date) %>%
    mutate(count = seq(n())) %>%
    ungroup() %>%
    arrange(date, desc(count))

winner_df %>% head(10)
```

```r
winner_df %>%
    head(10) %>%
    ggplot(aes(x = winner, y = count)) +
    geom_col() +
    coord_flip() +
    transition_time(date)
```
